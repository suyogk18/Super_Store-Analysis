```python
# for connecting jupyter notebook to poatgreSQL

import sqlalchemy as sal
from urllib.parse import quote_plus


USERNAME = "postgres"  
PASSWORD = quote_plus("Nevergiveup@2024")  
HOST = "localhost"  
PORT = "5432"  
DATABASE = "superstore_sql_project"

# Create engine
engine = sal.create_engine(f'postgresql://{USERNAME}:{PASSWORD}@{HOST}:{PORT}/{DATABASE}')

# Test connection
try:
    conn = engine.connect()
    print("Connected successfully!")
except Exception as e:
    print("Connection failed:", e)
```

```python
df.to_sql('store_record', con=conn, index=False, if_exists = 'replace')
```
